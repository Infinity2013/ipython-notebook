```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
data = pd.read_excel("../../data/2/FuelData2012.xlsx")
data.head()
```

```python
data.describe()
```

```python
data['Hwy MPG-cut'] = pd.cut(data['Hwy MPG'], bins=np.arange(15, 50, 5), right=False)
data.head()
```

```python
size_hwy_crosstab = pd.crosstab(data['Size'], data['Hwy MPG-cut'])
size_hwy_crosstab
```

```python
data['City MPG-cut'] = pd.cut(data['City MPG'], bins=np.arange(10, 50, 5), right=False)
data.head()
```

```python
drive_city_crosstab = pd.crosstab(data['Drive'], data['City MPG-cut'])
drive_city_crosstab
```

```python
fuel_city_crosstab = pd.crosstab(data['Fuel Type'], data['City MPG-cut'])
fuel_city_crosstab
```
