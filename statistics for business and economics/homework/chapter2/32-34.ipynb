```python
data.rename(columns={'Fund Name': 'name', 'Fund Type': 'type', 'Net Asset Value ($)': 'value', 
                      '5 Year Average Return (%)': 'return', 'Expense Ratio (%)': 'expense',
                     'Morningstar Rank': 'rank'}, inplace=True)
```

```python
data.head()
```

```python
data['return-cut'] = pd.cut(data['return'], bins=np.arange(0, 61, 10), right=False)
type_return_crosstab = pd.crosstab(data['type'], data['return-cut'])
type_return_crosstab
```

```python
type_per_crosstab = type_return_crosstab.copy()
for index in type_per_crosstab.index:
    type_per_crosstab.loc[index] = type_per_crosstab.loc[index] * 100 / type_per_crosstab.loc[index].sum()
type_per_crosstab
```

```python
return_perf_crosstab = type_return_crosstab.copy()
for column in return_perf_crosstab.columns:
    return_perf_crosstab[column] = return_perf_crosstab[column] * 100 / return_perf_crosstab[column].sum()
return_perf_crosstab
```

```python
data['expense-cut'] = pd.cut(data['expense'], bins=np.arange(0, 1.51, 0.25), right=False)
type_expense_crosstab = pd.crosstab(data['type'], data['expense-cut'])
type_expense_crosstab
```

```python
expense_freq_df = pd.DataFrame(data['expense-cut'].value_counts().to_dict().items(), columns=['name', 'freq'])
expense_freq_df.sort_values(by='name')
expense_freq_df['per'] = expense_freq_df['freq'] * 100 / expense_freq_df['freq'].sum()
expense_freq_df
```
