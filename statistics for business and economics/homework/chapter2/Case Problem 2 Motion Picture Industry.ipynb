```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
data = pd.read_excel('../../data/2/2011Movies.xlsx')
data.head()
```

```python
data.describe()
```

```python
data.sort_values(by='Total Gross Sales ($millions)',ascending=False).head()
```

```python
data.rename(columns={'Motion Picture': 'name', 'Opening Gross Sales ($millions)': 'opening', 
                    'Total Gross Sales ($millions)': 'total', 'Number of Theaters': 'theater',
                    'Weeks in Release': 'week'}, inplace=True)
```

```python
data.head()
```

```python
index = 1
plt.figure(figsize=(10, 10))
for column in data.columns:
    if column == 'name': continue
    arr = pd.cut(data[column], bins=5).value_counts().to_dict()
    axis = plt.subplot(2, 2, index)
    axis.set_title(column)
    
    plt.pie(arr.values(), labels=arr.keys(), autopct='%1.1f%%')
    
    

    index += 1
    
```

```python
index = 1
for column in ['opening', 'theater', 'week']:
    axis = plt.subplot(2, 2, index)
    axis.set_title('total - %s scatter' % column)
    plt.scatter(data[column], data['total'])
    index +=1 
plt.tight_layout()
```
