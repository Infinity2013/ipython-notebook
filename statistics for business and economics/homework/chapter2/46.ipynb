```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
data = pd.read_excel("../../data/2/2012Population.xlsx")
data.head()
```

```python
data.describe()
```

```python
data['population-cut'] = pd.cut(data['Population'], bins=np.arange(0, 40, 2.5), right=False)
populationDis = pd.DataFrame(data['population-cut'].value_counts().to_dict().items(), columns=['range', 'freq'])
populationDis['percentage'] = populationDis['freq'] * 100 / populationDis['freq'].sum()
populationDis.sort_values(by='range', inplace=True)
```

```python
left = np.arange(15)
plt.bar(left, populationDis['freq'], 0.3)
plt.xticks(left, populationDis['range'], rotation=45)
```
