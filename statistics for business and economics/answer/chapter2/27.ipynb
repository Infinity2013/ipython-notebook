```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
data = pd.read_excel('../../data/2/Crosstab.xlsx')
data
```

```python
orignCrosstab = pd.crosstab(data['x'], data['y'])
```

```python
percentCrosstab = orignCrosstab.copy()
```

```python
for index in percentCrosstab.index:
    percentCrosstab.loc[index] = percentCrosstab.loc[index] * 100 / percentCrosstab.loc[index].sum()
percentCrosstab
```

```python
percentCrosstab2 = orignCrosstab.copy()
```

```python
for column in percentCrosstab2.columns:
    percentCrosstab2[column] = percentCrosstab2[column] * 100 / percentCrosstab2[column].sum()
percentCrosstab2
    
```
